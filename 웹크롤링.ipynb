### library 크롤링

```python
import requests
from bs4 import BeautifulSoup

webpage = requests.get('https://library.gabia.com/')
soup = BeautifulSoup(webpage.content, 'html.parser')

elements = soup.select('div.esg-entry-content a> span')

for index, elements in enumerate(elements, 1):
    print("{}번째 게시글의 제목 : {}" .format(index, elements.text))
```

```python
import requests
from bs4 import BeautifulSoup

webpage = requests.get('https://library.gabia.com/')
soup = BeautifulSoup(webpage.content, 'html.parser')

elements = soup.select('div.esg-entry-content a.eg-grant-element-0')

for index, elements in enumerate(elements, 1):
    print("{}번째 게시글의 제목 : {}, {}" .format(index, elements.text, elements.attrs['href']))
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

webpage = requests.get('https://library.gabia.com/')
soup = BeautifulSoup(webpage.content, 'html.parser')

elements = soup.select('div.esg-entry-content a.eg-grant-element-0')

titles = []
links = []

for index, element in enumerate(elements, 1):
    titles.append(element.text)
    links.append(element.attrs['href'])
    
df = pd.DataFrame()
df['titles'] = titles
df['links'] = links
df.to_excel('./gabia.xlsx', sheet_name = "Sheet1")
```
